```python
Aim:
To explore and analyze a dataset using Python libraries such as Pandas, NumPy, and Matplotlib by performing data loading, filtering, statistical analysis, and visualization operations.
```

```python
Data Used:

Airline dataset given in the url and a randomly generated dataset representing property sales data

Attributes include:

Date

Property Price (in INR)

Number of Properties Sold

The dataset spans from January 2010 to December 2020
```

```python
Procedure:

Import required Python libraries such as Pandas, NumPy, and Matplotlib.

Create a random dataset using NumPy and store it in a Pandas DataFrame.

Display the first and last few records to understand the dataset structure.

Generate summary statistics to analyze data distribution.

Filter data for a specific year.

Compute average property price per year and visualize it using a line plot.

Count the total number of properties sold per year.

Query the dataset for a specific date range.

Calculate the mean property price on a month-wise basis using GroupBy.

Visualize the price distribution using a histogram.

Filter and display properties with prices greater than 5 Lakhs.
```

```python
# Experiment No: 01
# Load a time-series data, peek, query by date-time, and review statistics

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ==================================================
# DATASET 1: AIRLINE PASSENGERS DATASET
# ==================================================

# Step 1: Load dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)

# Step 2: Display first and last rows
print("First 5 rows:\n", df.head())
print("\nLast 5 rows:\n", df.tail())

# Step 3: Summary statistics
print("\nSummary Statistics:\n", df.describe())

# Convert Month to datetime
df['Month'] = pd.to_datetime(df['Month'])
df['Year'] = df['Month'].dt.year
df['Month_No'] = df['Month'].dt.month

# Step 4: Filter data for a specific year
print("\nData for Year 1955:\n", df[df['Year'] == 1955])

# Step 5: Plot average passengers per year
df.groupby('Year')['Passengers'].mean().plot(marker='o', title='Average Passengers Per Year')
plt.xlabel("Year")
plt.ylabel("Average Passengers")
plt.grid()
plt.show()

# Step 6: Count per year
print("\nCount per year:\n", df.groupby('Year').size())

# Step 7: Date range query
print("\nData from Jan 1952 to Dec 1956:\n",
      df[(df['Month'] >= '1952-01-01') & (df['Month'] <= '1956-12-31')])

# Step 8: Mean passengers month-wise
print("\nMean Passengers Month-wise:\n",
      df.groupby('Month_No')['Passengers'].mean())

# Step 9: Histogram
plt.hist(df['Passengers'], bins=10, edgecolor='black')
plt.title("Passengers Histogram")
plt.xlabel("Passengers")
plt.ylabel("Frequency")
plt.show()

# Step 10: Passengers > 500
print("\nPassengers greater than 500:\n", df[df['Passengers'] > 500])


# ==================================================
# DATASET 2: RAW SALES DATASET
# ==================================================

# Step 1: Load raw sales dataset (file must be in same folder)
sales_df = pd.read_csv("raw_sales.csv")

# Normalize column names
sales_df.columns = sales_df.columns.str.lower().str.strip().str.replace(" ", "_")

print("\nSales Dataset Columns:", sales_df.columns)

# Convert datesold to datetime
sales_df['datesold'] = pd.to_datetime(sales_df['datesold'])

# Ensure price is numeric
sales_df['price'] = pd.to_numeric(sales_df['price'], errors='coerce')

# Extract year and month
sales_df['year'] = sales_df['datesold'].dt.year
sales_df['month'] = sales_df['datesold'].dt.month

# Step 2: Display first and last rows
print("\nSales Dataset Head:\n", sales_df.head())
print("\nSales Dataset Tail:\n", sales_df.tail())

# Step 3: Summary statistics
print("\nSales Summary Statistics:\n", sales_df.describe())

# Step 4: Filter data for a specific year
print("\nSales Data for Year 2015:\n", sales_df[sales_df['year'] == 2015])

# Step 5: Average price per year
sales_df.groupby('year')['price'].mean().plot(marker='o', title='Average Sales Price Per Year')
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.grid()
plt.show()

# Step 6: Count of properties sold per year
print("\nProperties sold per year:\n", sales_df.groupby('year').size())

# Step 7: Date range query
print("\nSales from Jan 2010 to Dec 2015:\n",
      sales_df[(sales_df['datesold'] >= '2010-01-01') &
               (sales_df['datesold'] <= '2015-12-31')])

# Step 8: Mean price month-wise
print("\nMean Sales Price Month-wise:\n",
      sales_df.groupby('month')['price'].mean())

# Step 9: Histogram
plt.hist(sales_df['price'], bins=15, edgecolor='black')
plt.title("Sales Price Histogram")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# Step 10: Property price > 5 Lakhs
print("\nProperty Prices > 5 Lakhs:\n", sales_df[sales_df['price'] > 500000])
```

```python
# Experiment No: 01
# Randomly Generated Dataset and Operations

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# STEP 1: CREATE RANDOM DATASET
# ==============================

np.random.seed(10)

dates = pd.date_range(start='2010-01-01', end='2020-12-31', freq='M')
prices = np.random.randint(100000, 1000000, size=len(dates))  # Price in INR
properties_sold = np.random.randint(5, 50, size=len(dates))

df = pd.DataFrame({
    'Date': dates,
    'Price': prices,
    'Properties_Sold': properties_sold
})

df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# =====================================
# STEP 2: DISPLAY FIRST & LAST FEW ROWS
# =====================================

print("First 5 rows:\n", df.head())
print("\nLast 5 rows:\n", df.tail())

# =========================
# STEP 3: SUMMARY STATISTICS
# =========================

print("\nSummary Statistics:\n", df.describe())

# =====================================
# STEP 4: FILTER DATA FOR SPECIFIC YEAR
# =====================================

print("\nData for Year 2015:\n", df[df['Year'] == 2015])

# =====================================
# STEP 5: PLOT AVERAGE PRICE PER YEAR
# =====================================

df.groupby('Year')['Price'].mean().plot(marker='o', title='Average Property Price Per Year')
plt.xlabel("Year")
plt.ylabel("Average Price")
plt.grid()
plt.show()

# =====================================
# STEP 6: COUNT OF PROPERTIES SOLD PER YEAR
# =====================================

print("\nTotal Properties Sold Per Year:\n",
      df.groupby('Year')['Properties_Sold'].sum())

# =====================================
# STEP 7: QUERY SPECIFIC DATE RANGE
# =====================================

print("\nData from Jan 2010 to Dec 2015:\n",
      df[(df['Date'] >= '2010-01-01') & (df['Date'] <= '2015-12-31')])

# =====================================
# STEP 8: MEAN PRICE MONTH-WISE (GROUPBY)
# =====================================

print("\nMean Price Month-wise:\n",
      df.groupby('Month')['Price'].mean())

# =====================================
# STEP 9: HISTOGRAM PLOT
# =====================================

plt.hist(df['Price'], bins=15, edgecolor='black')
plt.title("Histogram of Property Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()

# =====================================
# STEP 10: PROPERTY PRICE > 5 LAKHS
# =====================================

print("\nProperties with Price > 5 Lakhs:\n", df[df['Price'] > 500000])
```
